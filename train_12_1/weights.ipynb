```python
def findMaxandMin(ls):
    Max = -10 #tang nhieu nhat
    Min = 10 #giam nhieu nhat
    for i in range(1, len(ls)):
        delta = ls[i] - ls[i - 1]
        if delta > 0 and delta > Max:
            Max = delta 
        if delta < 0 and delta < Min:
            Min = delta 
    return Max, Min
```

```python
def countDeandIn(ls):
    de = 0
    inc = 0
    for i in range(1, len(ls)):
        if ls[i] > ls[i - 1]:
            inc += 1
        else:
            de += 1
    return de, inc
```

```python
import matplotlib.pyplot as plt

w = [
    84.3, 83.2, 82.8, 82.0, 82.7
]
d = [
    '13/1 (night)', '14/1 (day)', '14/1 (night)', '15/1 (day)', '15/1 (night)'
]

plt.plot(d, w, marker='.')
plt.xlabel('Time')
plt.ylabel('Weights (kg)')
plt.title('Weights by day')

#oneday has 2 parameters, 0:00 and after wake up

print(f'Max in process: {max(w)}')
print(f'Min in process: {min(w)}')

maxIncrease, maxDecrease = findMaxandMin(w)

print(f'Max increase: {maxIncrease}')
print(f'Max decrease: {maxDecrease}')

plt.show()
```

```python
de, inc = countDeandIn(w)

de_percent = 100 * (round(de / (de + inc), 2))

inc_percent = 100 - de_percent

print(f'Number of decrease process: {de} - {de_percent}%')
print(f'Number of increase process: {inc} - {inc_percent}%')
columns = ['Decrease', 'Increase']
values = [de, inc]
plt.bar(columns, values)
plt.xlabel('Count')
plt.ylabel('Number of process')
plt.show()

labels = ['decrease', 'increase']
sizes = [de, inc]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```
