```python
import matplotlib.pyplot as plt

w = [84.3, 83.2]
d = ['13/1 (night)', '14/1 (day)']

plt.plot(d, w, marker='.')
plt.xlabel('Date')
plt.ylabel('Weights (kg)')
plt.title('Weights by day')

#oneday has 2 parameters, 0:00 and after wake up

print(f'Max in process: {max(w)}')
print(f'Min in process: {min(w)}')

plt.show()
```
