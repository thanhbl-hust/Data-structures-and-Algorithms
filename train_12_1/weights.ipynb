```python
def findMaxandMin(ls):
    Max = -10 #tang nhieu nhat
    Min = 10 #giam nhieu nhat
    for i in range(1, len(ls)):
        delta = ls[i] - ls[i - 1]
        if delta > 0 and delta > Max:
            Max = delta 
        if delta < 0 and delta < Min:
            Min = delta 
    return Max, Min
```

```python
def countDeandIn(ls):
    de = 0
    inc = 0
    for i in range(1, len(ls)):
        if ls[i] > ls[i - 1]:
            inc += 1
        else:
            de += 1
    return de, inc
```

```python
import matplotlib.pyplot as plt

w = [
    84.3, 83.2, 82.8, 82.0, 82.7, 81.6, 82.6, 81.5,
    82, 81.5
]
d = [
    '13/1 N', '14/1 D', '14/1 N', '15/1 D', '15/1 N', '16/1 D', '16/1 N', '17/1 D',
    '17/1 N', '18/1 D'
]


print(len(w))
print(len(d))
plt.plot(d, w, marker='.')
plt.xlabel('Time')
plt.ylabel('Weights (kg)')
plt.title('Weights by day')

#oneday has 2 parameters, 0:00 and after wake up

print(f'Max weight in process: {round(max(w), 2)} (kg)')
print(f'Min weight in process: {round(min(w), 2)} (kg)')

maxIncrease, maxDecrease = findMaxandMin(w)

print(f'Max weight increase in one process: {round(maxIncrease, 2)} (kg)')
print(f'Max weight decrease in one process: {-1 * round(maxDecrease, 2)} (kg)')

plt.show()
```

```python
de, inc = countDeandIn(w)

de_percent = 100 * (round(de / (de + inc), 3))

inc_percent = 100 - de_percent

print(f'Number of decrease process: {de} - {de_percent}%')
print(f'Number of increase process: {inc} - {inc_percent}%')
columns = ['Decrease', 'Increase']
values = [de, inc]
plt.bar(columns, values)
plt.xlabel('Count de and in process')
plt.ylabel('Number of process')
plt.show()

labels = ['decrease', 'increase']
sizes = [de, inc]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.xlabel('De and in in percent')
plt.show()
```
